## worldometer_data.csv

```python
# Replace missing values in these columns with 0
df6.fillna({'NewCases':0, 'NewDeaths':0, 'NewRecovered':0}, inplace=True)
df6.isnull().sum()
```

```python
# Fill missing values with 'Unkown'
df6['WHO Region'] = df6['WHO Region'].fillna('Unknown')
```

```python
# Drop rows with missing values
df6.dropna(inplace=True)
```

## Line chart for daily new cases, deaths, and recoveries over time

```python
time_series_data = df2.groupby('Date').sum()

plt.figure(figsize=(12, 6))
plt.plot(time_series_data.index, time_series_data['Confirmed'], label='Total Cases')
plt.plot(time_series_data.index, time_series_data['Deaths'], label='Total Deaths')
plt.plot(time_series_data.index, time_series_data['Recovered'], label='Total Recovered')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Time Series Analysis')
plt.legend()
plt.grid(True)
plt.show()
```

## Total COVID-19 Deaths by Continent

```python
import seaborn as sns
plt.figure(figsize=(12,6))
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.title('Total Deaths by Continent')
gathered_global_death = df2.groupby('WHO Region')['Deaths'].sum().reset_index().sort_values(by='Deaths', ascending=False)
sns.barplot(data=gathered_global_death, x='WHO Region', y='Deaths',palette="viridis")
plt.show()
```

## Top 10 countries that has been given maximum number of covid cases

```python
top_countries_confirmed = df1.groupby('Country/Region')['Confirmed'].max().nlargest(10)
colors = ['skyblue', 'lightcoral', 'lightgreen', 'orange', 'pink', 'yellow', 'violet', 'gold', 'lightblue', 'salmon']
plt.figure(figsize=(10,6))
top_countries_confirmed.plot(kind='bar', color=colors)
plt.xlabel('Country/Region')
plt.xlabel("Country/Region")
plt.ylabel("Total Confirmed Cases")
plt.title("Countries with Most Cases")
plt.xticks(rotation =45)
plt.tight_layout()
plt.show()
```

## Ten Countries with Most Deaths from COVID-19

```python
plt.figure(figsize=(12,6))
countries_most_deaths = df2.groupby(['Country/Region'])[['Deaths']].sum().reset_index().sort_values(by='Deaths', ascending=False).head(10).reset_index(drop=True)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.xticks(rotation=45)
plt.title('Countries with Most Deaths')
sns.barplot(data=countries_most_deaths, x='Country/Region', y='Deaths', palette="viridis")
plt.show()

```

## Global COVID-19 Death Percentage

```python
colors = ['lightcoral', 'lightskyblue', 'lightgreen', 'gold', 'orange', 'lightpink']
plt.figure(figsize=(8, 8))
plt.title('Global Death Percentage', pad=20, fontdict={'fontsize': 16, 'fontweight': 'bold'})
plt.pie(gathered_global_death['Deaths'],
        labels=[f"{region} ({deaths} deaths)" for region, deaths in zip(gathered_global_death['WHO Region'], gathered_global_death['Deaths'])],
        autopct='%1.1f%%',
        startangle=45,
        colors=colors,
        wedgeprops={'edgecolor': 'white', 'linewidth': 0})
plt.axis('equal')
plt.show()
```
